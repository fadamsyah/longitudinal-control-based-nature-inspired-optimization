```python
import numpy as np
import scienceplots
import matplotlib.pyplot as plt
import pandas as pd
from copy import deepcopy
from os import listdir
from os.path import isfile, join

plt.style.use(['science', 'grid', 'no-latex'])

DPI = 500
```

```python
def get_array_from_time_interval(x, t, T):
    t0 = min(T)
    t1 = max(T)
    i1 = -1
    i0 = 0
    for i in range(len(t)):
        if t0 <= t[i]:
            i0 = deepcopy(i)
            break
    for i in range(i0, len(t)):
        if t1 <= t[i]:
            i1 = deepcopy(i)
            break

    return x[i0:i1], t[i0:i1]
```

# Load Variables

```python
CSV0 = 'cost0/carla_sim.csv'
CSV1 = 'cost1/lamda_0_10/carla_sim.csv'
CSV2 = 'cost2/lamda_0_20/carla_sim.csv'
CSV3 = 'cost3/lamda_0_10/carla_sim.csv'

CSVs = [CSV0, CSV1, CSV2, CSV3]

for i, file in enumerate(CSVs):
    df = pd.read_csv(file)
    exec(f"t{i} = df['time'].values")
    exec(f"sp{i} = df['sp_v'].values")
    exec(f"gt{i} = df['speed'].values")
    exec(f"tht{i} = df['throttle'].values")
```

# Plot

##

```python
T = [0, 4.5]
NAME = 'r1'

l0, lt0 = get_array_from_time_interval(gt0, t0, T)
l1, lt1 = get_array_from_time_interval(gt1, t1, T)
l2, lt2 = get_array_from_time_interval(gt2, t2, T)
l3, lt3 = get_array_from_time_interval(gt3, t3, T)

spt0, lt0 = get_array_from_time_interval(sp0, t0, T)
spt1, lt1 = get_array_from_time_interval(sp1, t1, T)
spt2, lt2 = get_array_from_time_interval(sp2, t2, T)
spt3, lt3 = get_array_from_time_interval(sp3, t3, T)

sp = (spt0 + spt1 + spt2+ spt3) / 4

th0, _ =  get_array_from_time_interval(tht0, t0, T)
th1, _ =  get_array_from_time_interval(tht1, t1, T)
th2, _ =  get_array_from_time_interval(tht2, t2, T)
th3, _ =  get_array_from_time_interval(tht3, t3, T)
```

```python
plt.plot(lt1, sp, color='black', linestyle='--')
plt.plot(lt0, l0, label=r'Unregularized', zorder=1)
plt.plot(lt1, l1, label=r'Type 1 $\lambda=10$', zorder=2)
plt.plot(lt2, l2, label=r'Type 2 $\lambda=20$', zorder=3)
plt.plot(lt3, l3, label=r'Type 3 $\lambda=10$', zorder=4)

plt.xlabel('Time $(s)$')
plt.ylabel('Speed $(m/s)$')
plt.legend(fontsize=7)
plt.savefig(f'img/{NAME}.png', dpi=DPI, transparent=True, bbox_inches='tight')
plt.show()
```

```python
plt.plot(lt0, th0, label=r'Unregularized', zorder=1)
plt.plot(lt1, th1, label=r'Type 1 $\lambda=10$', zorder=1)
plt.plot(lt2, th2, label=r'Type 2 $\lambda=20$', zorder=1)
plt.plot(lt3, th3, label=r'Type 3 $\lambda=10$', zorder=1)

plt.xlabel('Time $(s)$')
plt.ylabel('Throttle')
plt.ylim([0,None])
plt.legend(fontsize=7)
plt.savefig(f'img/{NAME}_throttle.png', dpi=DPI, transparent=True, bbox_inches='tight')
plt.show()
```
